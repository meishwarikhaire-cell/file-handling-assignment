```python
# Task 1: Create attendance file

with open("attendance.txt", "w") as file:
    file.write("Rahul,Present,2024-01-15\n")
    file.write("Aisha,Absent,2024-01-15\n")
    file.write("Rohan,Present,2024-01-15\n")

print("attendance.txt file created successfully")
```

```python
# Show attendance.txt content
with open("attendance.txt","r") as file:
    print(file.read())
```

```python
# Task 2: Append new records

with open("attendance.txt", "a") as file:
    file.write("Neha,Present,2024-01-15\n")
    file.write("Vikram,Absent,2024-01-15\n")

print("New records added successfully")
```

```python
# Function to count present students

def count_present_students(filename):

    count = 0

    with open(filename, "r") as file:
        for line in file:
            parts = line.strip().split(",")

            if parts[1] == "Present":
                count += 1

    return count


total = count_present_students("attendance.txt")

print("Total students present:", total)
```

#Difference between 'r', 'w', and 'a'
'r' (Read Mode)
Used to read data from a file. The file must exist.

'w' (Write Mode)
Used to create a new file or overwrite existing content.

'a' (Append Mode)
Used to add new data at the end of the file without deleting existing content.

#Why we use with statement
The 'with' statement is used to handle files safely.
It automatically closes the file after the operation is completed.
This prevents memory leaks and makes the code cleaner.

#How JSON relates to Python dictionaries
JSON is similar to Python dictionaries because both store data in key-value pairs.

Example:

Python Dictionary
{"name": "Rahul", "marks": 85}

JSON
{
  "name": "Rahul",
  "marks": 85
}
